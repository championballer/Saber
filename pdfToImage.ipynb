```python
def convert_to_array(modif):
    result = [];
    for i in range(modif.shape[0]):
        sub_result=[]
        for j in range(modif.shape[1]):
            if modif[i][j]!='nan':
                sub_result.append(modif[i][j])
                
        if len(sub_result)!=0:
            result.append(np.array(sub_result))
        
    return np.array(result)
```

```python
import requests
```

```python
def pdfToTable(PDFfilename, apiKey, fileExt, downloadDir):
    fileData = (PDFfilename, open(PDFfilename, 'rb'))
    files = {'f': fileData}
    postUrl = "https://pdftables.com/api?key={0}&format={1}".format(apiKey, fileExt)
    response = requests.post(postUrl, files=files)
    response.raise_for_status()
    with open(downloadDir, "wb") as f:
        f.write(response.content)
```
